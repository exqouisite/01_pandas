##5.3 Handling Missing Link

###5.3.1 Remove The Row of Missing Value


```python
#Remove missing value
df_without_nan = df.dropna()

print(df_without_nan)
```

###5.3.2 Finding The Missing Value

```python
#1. Menggunakan Min atau Max
fill_by_max = df['Ranks'].fillna(df['Ranks'].max())
fill_by_min = df['Ranks'].fillna(df['Ranks'].min())

print (fill_by_max)
print (fill_by_min)
```

```python
#2. Menggunakan Mean
fill_by_mean = df['Marks'].fillna(df['Marks'].mean())
fill_by_median = df['Marks'].fillna(df['Marks'].median())

print (fill_by_mean)
print (fill_by_median)
```

```python
#3. Menggunakan Median

fill_by_median = df['Marks'].fillna(df['Marks'].median())

print (fill_by_median)
```

```python
#Menggunakan mode (frequent item)
fill_by_mode = df["Ranks"].fillna(df["Ranks"].mode().iloc[0])

print (fill_by_mode)
print(f"How many null now = {fill_by_mode.isnull().sum()}")
```

```python
#Menggunakan Backward atau Forward
fill_by_bfill = df['Ranks'].fillna(method='bfill')
fill_by_ffill = df['Ranks'].fillna(method='ffill')

print (fill_by_bfill)
print (fill_by_ffill)
```

```python
#Menggunakan Interpolation Linear
fill_by_interpolate_linear = df["Ranks"].interpolate(method='linear')
print (fill_by_interpolate_linear)

#Menggunakan Interpolation Quadratic
fill_by_interpolate_quadratic = df["Ranks"].interpolate(method='quadratic')
print (fill_by_interpolate_quadratic)

#Menggunakan Interpolation Spline
fill_by_interpolate_spline = df["Ranks"].interpolate(method='spline', order=2)
print (fill_by_interpolate_spline)
```
